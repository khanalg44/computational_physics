# Bouncing ball

Finding the motion of an elastic object: We model a onedimensional elastically deforming ball as a set of masses, m, connected by springs of spring constant k. The springs are enumerated $i = 1, . . . , N$ from the bottom to the top. The bottom-most mass also interacts with the floor. The force on particle i is then 


$$m_i a_i = F_i = −m_i g − k (z_{i+1} − z_i − b) + k (z_i − z_{i−1}) − F_w$$

where $F_w = −k_w z_i$ when for i = 1 when $z_i < 0$ and 0 otherwise.
The motion is solved using the following program:

```python
import numpy as np
import pylab as plt
```

```python
k = 100000
kw = 1000000 # Spring constants
g = 9.81 # Acc of gravity


def BouncingBall():    
    N = 10 # Nr of elements    
    d = 0.1 # Size in m
    m = 0.1 # Mass in kg
    t = 2.0 # Simulation time
    z0 = 0.1 # Initial height
    dt = 0.0001 # Timestep
    nstep = int(np.ceil(t/dt))
    
    print (nstep)
    
    z = np.zeros((N,nstep),float)
    vz = np.zeros((N,nstep),float)
    z[:,0] = z0 + d*(np.array(range(N)))/N # Initial positions
    l = d/N # Distance between nodes
    for i in range(nstep-1): # Calculate motion
        dz = diff(z[:,i])-l
        F = -k*append(0.0,dz) + k*append(dz,0.0) - m*g # Internode forces
        F[0] = F[0] - kw*z[0,i]*(z[0,i]<0) # Bottom wall
        a = F/m
        vz[:,i+1] = vz[:,i] + a*dt
        z[:,i+1] = z[:,i] + vz[:,i+1]*dt;
    return z


z= BouncingBall()
print (z.shape)


#for i in range(1):
#    plt.plot(range(z.shape[1]), z[i,:])

plt.plot(range(z.shape[1]), z[0,:])
plt.plot(range(z.shape[1]), z[1,:]);
plt.plot(range(z.shape[1]), z[2,:]);
```
